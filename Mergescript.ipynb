```python

def getSeason(month):
    if (month == 12 or month == 1 or month == 2):
       return "WINTER"
    elif(month == 3 or month == 4 or month == 5):
       return "SPRING"
    elif(month ==6 or month==7 or month ==7 or month ==8):
       return "SUMMER"
    else:
       return "FALL"
# d1['season'].head()
```
